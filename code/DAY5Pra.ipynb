```python
#str.strip()
"""
str.strip() 结论：
1. 默认参数：不传入参数时，默认删除字符串首尾的空白字符
2. 空白字符包括：空格(' ')、制表符('\t')、换行符('\n')、回车符('\r')、换页符('\f')等
3. 只删除首尾，不删除中间的空格
4. 返回新字符串，不修改原字符串
5. 不会删除非空白字符（如"-"、"*"等）
"""

class Shape:
    count=0
    def __init__(self):
        self._area=0
        Shape.count+=1


    def _calc_area(self):
        return 0

    def get_area(self):
        self._area=self._calc_area()
        return self._area


    @classmethod
    def get_count(cls):
        return cls.count


class Circle(Shape):
    def __init__(self,radius):
        super().__init__()
        self.radius=radius

    def _calc_area(self):
        import math
        return round(math.pi*self.radius**2,2)

class Rectangle(Shape):
    def __init__(self,width,height):
        super().__init__()
        self.width=width
        self.height=height

    def _calc_area(self):
        return self.width*self.height


def main():
    circle_1=Circle(2)
    rectangle_1=Rectangle(7,9)

    print(f'the area of circle_1:{circle_1.get_area()}' )
    print(f'the area of rectangle_1:{rectangle_1.get_area()}' )
    print(f'')
    


if __name__=="__main__":
    main()
    
```
