```python
#****要求会推导计算一个神经网络前向传播的公式****
"""
直接计算：
输入: x₁, x₂, ..., xₙ
权重: w₁, w₂, ..., wₙ
偏置: b
加权和: z = w₁x₁ + w₂x₂ + ... + wₙxₙ + b
激活函数: a = σ(z) 


向量形式: 
z = W·X + b = WᵀX + b
a = σ(z)
"""

#****用直接计算，和矩阵的方式，实现一个神经元****

import numpy as np

#定义激活函数：
def sigmoid(x):
    return 1/(1+np.exp(-x))


def relu(x):
    return np.maximum(0,x)



#定义神经元类：
class Neuron():
    def __init__(self,dim_input):
        self.weights = np.random.randn(dim_input) * 0.1
        self.bias = np.random.randn() * 0.1


#直接计算：
    def forward_direct(self,x):
        z = self.bias
        for i in range(len(x)):
            z += self.weights[i]*x[i]

        return sigmoid(z)

#单项计算：
    def forward_single(self,x):
        z = np.dot(self.weights,x) + self.bias

        return sigmoid(z)

#批量计算：
    def forward_batch(self,X):
        Z = np.dot(X,self.weights) + self.bias

        return sigmoid(Z)


#固定随机数：
np.random.seed(11)

X_single=np.array([0.5,0.7,-0.4])
X_batch=np.random.randn(4,3)

neuron = Neuron(dim_input=3)

output_direct = neuron.forward_direct(X_single)
output_single = neuron.forward_single(X_single)
output_batch = neuron.forward_batch(X_batch)

print(f"单个样本输入: {X_single}")
print(f"直接计算结果: {output_direct:.5f}")
print(f"向量方式结果: {output_single:.5f}")
print(f"\n批量处理4个样本的结果:\n{output_batch}")



#****将上述神经元组成网络，实现前向传播****


"""
权重矩阵的维度 = (上一层输出维度，下一层神经元数)    

偏置向量的维度 = (下一层神经元数)

input_dim: 输入层的 “特征数”（每个样本有几个输入值）

hidden_dim: 隐藏层的 “神经元个数”（这一层有多少个神经元）

output_dim: 输出层的 “结果数”（最终要输出几个值）
"""

class NeuralNetwork:
    def __init__(self, input_dim, hidden_dim, output_dim):
        # 隐藏层参数：W1形状=(输入维度, 隐藏层神经元数)
        self.W1 = np.random.randn(input_dim, hidden_dim) * 0.1
        self.b1 = np.random.randn(hidden_dim) * 0.1 
        
        # 输出层参数：W2形状=(隐藏层神经元数, 输出维度)
        self.W2 = np.random.randn(hidden_dim, output_dim) * 0.1
        self.b2 = np.random.randn(output_dim) * 0.1
    
    def forward(self, X):
       
        # 第一步：输入层→隐藏层
        Z1 = np.dot(X, self.W1) + self.b1
        #隐藏层输出
        A1 = sigmoid(Z1)
        
        # 第二步：隐藏层→输出层
        Z2 = np.dot(A1, self.W2) + self.b2
        # 最终输出
        A2 = sigmoid(Z2)
        
        # 返回各层结果
        return Z1, A1, Z2, A2

#测试神经网络前向传播
if __name__ == "__main__":
    # 固定随机数
    np.random.seed(11)
    
    #输入数据
    X_single = np.array([0.5, 0.7, -0.4])  
    X_batch = np.random.randn(4, 3)      
    
    #创建神经网络：3维输入 → 4个隐藏层神经元 → 1维输出
    nn = NeuralNetwork(input_dim=3, hidden_dim=4, output_dim=1)
    
    # 单个样本前向传播
    Z1, A1, Z2, A2_single = nn.forward(X_single)
    print("\n单个样本前向传播结果：")
    print(f"隐藏层加权和(Z1)：{Z1}")
    print(f"隐藏层输出(A1)：{A1}")
    print(f"最终输出(A2)：{A2_single[0]:.5f}")  # 单个样本取第一个元素
    
    # 批量样本前向传播
    _, _, _, A2_batch = nn.forward(X_batch)
    print("\n批量样本最终输出：")
    print(A2_batch)
```
